```python
import cv2
import numpy as np
import os
```

```python
# this bit is to read and capture the saturation and hue of any general piece of cloth
cam = cv2.VideoCapture(0)
ret, frame = cam.read()
frame.shape
hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

hue_midpt = hsv_frame[int(hsv_frame.shape[0]/2), int(hsv_frame.shape[1]/2), 0]
thresh_left = hue_midpt-20
thresh_right = hue_midpt+20
```

```python
cam = cv2.VideoCapture(0)
ret, orig_frame = cam.read()
hsv_frame_orig = cv2.cvtColor(orig_frame, cv2.COLOR_BGR2HSV)

# Get frame size
frame_size = (orig_frame.shape[1], orig_frame.shape[0])  # (width, height)


save_path = r"C:\Users\Asghar\cloak_output.mp4"

# VideoWriter
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  
fps = 20.0
out = cv2.VideoWriter(save_path, fourcc, fps, frame_size)

print("Saving to:", os.path.abspath(save_path))  # For confirmation in terminal[web:39][web:41]

while True:
    rval, frame = cam.read()
    if not rval:
        break
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    #    frame[hsv_frame[:,:,0] > 80 and hsv_frame[:,:,0] < 200] = orig_frame
#     mask_single_channel = (hsv_frame[:,:,0] > 80)*(hsv_frame[:,:,0] <200)
#     mask_mkc = np.array([mask_single_channel, mask_single_channel, mask_single_channel])
#     mask_mkc = mask_mkc.transpose((1, 2, 0))
#    mask = (hsv_frame[:,:,0] > thresh_left) * (hsv_frame[:,:,0] < thresh_right)
#    mask = (hsv_frame[:,:,2] > 50) * (hsv_frame[:,:,0] < 128)
    mask = (hsv_frame[:,:,2] > 150) * (hsv_frame[:,:,1] < 40)
    
    new_array = orig_frame
    frame[mask] = new_array[mask]
    
    # Write processed frame to video
    out.write(frame)
    
    cv2.imshow("feed", frame)
    
    key = cv2.waitKey(20)
    if key == 27:  # exit on ESC
        break

cv2.destroyAllWindows()
cam.release()
out.release()  # Crucial: Releases and finalizes video file
```
